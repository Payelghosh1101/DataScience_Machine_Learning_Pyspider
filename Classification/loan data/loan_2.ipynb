```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import precision_score,recall_score,f1_score,accuracy_score,confusion_matrix
```

```python
df=pd.read_csv("loan_data.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
x=df.drop('loan_status',axis=1)
y=df['loan_status']
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
num_cols= xtrain.select_dtypes(exclude="object").columns
cat_cols=xtrain.select_dtypes(include="object").columns
```

```python
scaler=StandardScaler()
encode=OneHotEncoder()
```

```python
xtrain[num_cols]=scaler.fit_transform(xtrain[num_cols])
xtest[num_cols]=scaler.transform(xtest[num_cols])
```

```python
xtrain
```

```python
encoded_xtrain=encode.fit_transform(xtrain[cat_cols])
xtrain[encode.get_feature_names_out()]=encoded_xtrain.toarray()
```

```python
encode.get_feature_names_out()
```

```python
encoded_xtest=encode.transform(xtest[cat_cols])
xtest[encode.get_feature_names_out()]=encoded_xtest.toarray()
```

```python
xtrain = xtrain.drop(columns=cat_cols)
xtest = xtest.drop(columns=cat_cols)
```

```python
xtrain_cat_df=pd.DataFrame(encoded_xtrain.toarray(),columns=encode.get_feature_names_out())
xtest_cat_df=pd.DataFrame(encoded_xtest.toarray(),columns=encode.get_feature_names_out())
```

```python
xtrain_cat_df
```

```python
xtrain_concat=pd.concat([xtrain.reset_index(drop=True),xtrain_cat_df.reset_index(drop=True)],axis=1)
xtest_concat=pd.concat([xtest.reset_index(drop=True),xtest_cat_df.reset_index(drop=True)],axis=1)
```

```python
xtrain_concat
```

```python
xtest_concat
```

```python
model=LogisticRegression()
```

```python
model.fit(xtrain_concat,ytrain)
```

```python
ypred=model.predict(xtest_concat)
```

```python
model.score(xtest_concat,ytest)
```

## Model evaluation

```python
confusion_matrix(ytest,ypred)
```

```python
precision_score(ytest,ypred,pos_label=1)
```

** Out of all the predictions approved loans (Class 1) was 0.775. Meaning 77% approved predictions were Truely approved.

```python
precision_score(ytest,ypred,pos_label=0)
```

** Out of all the predictions labeled as "Loan Not Approved", 92% were truly not approved.

```python
recall_score(ytest,ypred,pos_label=1)
```

** Out of all the actual "Loan Approved", 74% were Truely approved

```python
recall_score(ytest,ypred,pos_label=0)
```

**Out of all the actual "Loan Not Approved", 93% were Truely not approved

```python
f1_score(ytest,ypred,pos_label=1) 
```

** The F1-score for this class 1(Loan Approved) is 75%, a good balance between precision and recall

```python
f1_score(ytest,ypred,pos_label=0) 
```

** The F1-score for this class 0(Loan Not Approved) is 93%, showing strong balance between precision and recall
